#### Q3. US - Babyname dataset
#### 코드블록 내에 주어진 문제에 맞게 코드를 작성하시오.


```python
#Q3 US-baby dataset
#Q3-0. 아래 url을 통해 파일을 읽으시오.

import pandas as pd
baby_names = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')

baby_names
```

```python
#Q3-1. 첫 10 행을 출력하시오.
# head 함수를 이용하여 첫10 행을 출력
baby_names.head(10)
```

```python
#Q3-2. 'Unnamed: 0'과 'Id' 컬럼을 삭제하시오.
# drop 함수를 이용하여 컬럼 삭제
# 제거할 열의 이름과 열 방향 혹은 행 방향으로 진행할 것인지 작성
# axix = 1 -> 열방향
baby_names.drop(['Unnamed: 0', 'Id'], axis=1)
```

```python
#Q3-3. Count 순으로 정렬하시오.
# Count 값으로 정렬
# ascending = True(오름차순) , False(내림차순)
baby_names.sort_values('Count', ascending = True)
```

```python
#Q3-4. Year 컬럼을 삭제하고, Name을 기준으로 grouping하여 sum하시오. (groupby 사용. df.groupby('index_name'))
# drop 함수를 이용하여 Year 열 삭제
# groupby 함수를 이용하여 Name 기준으로 grouping -> 이후 sum함수를 사용
# numeric_only=False -> 추가하지 않으면 경고메세지 출력이 됨..
baby_names.drop('Year', axis=1).groupby('Name').sum()
```

```python
#Q3-5. Q3-4에서 Count를 기반으로 내림차순으로 정렬하여 가장 많은 빈도를 가지는 이름 5개를 출력하시오.
# Count를 기준으로 내림차순 정렬
# 이후 head 함수를 이용하여 5개의 행을 출력
baby_names.sort_values('Count', ascending = True)
baby_names.head(5)

#Q3 End
```

#### Q4. Chipotle dataset
#### 아래는 치폴레(chipotle) 판매 기록을 나타낸 dataset이다. 가장 많이 판매된 item 5개를 matplotlib의 bar()를 통해 오름차순으로 가시화하시오.
#### 빈도수를 측정할 때 Collection 패키지의 Counter를 사용해도 됨.


```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

chipotle_url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(chipotle_url, sep = '\t')

data = chipo.groupby('item_name').sum().sort_values(by=['quantity'], ascending = False).head(5)
data = data.drop("order_id", axis = 1)

# bar그래프 생성
plt.figure(figsize=(10, 6))
# x축은 아이템의 이름, y축은 아이템의 개수
data.plot.bar(rot = 0)
plt.xlabel('Items')
plt.ylabel('Number of Times Ordered')
plt.title('Most ordered Chipotle\'s Items')
plt.figure(figsize=(20, 6))
plt.show()

#Q4 End
```

#### Q5. Titanic dataset
#### Titanic 탑승객의 남/녀 성비를 pie()를 통해 시각화하시오. (Males=blue, Females=red)

```python
#Q5 Start
import pandas as pd
import matplotlib.pyplot as plt

titmic_url= 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv'
titanic = pd.read_csv(titmic_url)

data = titanic.groupby("Sex").count().loc[:, "PassengerId"]
Sex = ['Females', 'male']

plt.figure(figsize=(10, 6))
plt.pie(data, explode=(0, 0), labels=Sex, autopct='%1.1f%%', shadow=True, startangle=90, colors=['red', 'blue'])
plt.title("Sex Proportion")
plt.show()

#Q5 End
```
