#### Q1. sin(x), cos(x), tan(x)를 matplotlib을 이용해 시각화하시오.
##### 1. subplots()를 이용해 각 함수를 따로 시각화한다.
##### 2. x의 범위는 코드 내 예시를 활용하여 적절한 값을 사용한다.
##### 3. sin, cos, tan 함수는 math library를 활용하여 작성한다.
##### 4. 아래 예시와 같이 label을 지정하여 출력한다.
##### 예시:  
![nn](fig_01.png)

```python
#Q1 Start
%matplotlib inline
import math
import matplotlib.pyplot as plt
from matplotlib import rc

# 폰트의 크기를 설정하기 위한 딕셔너리
font = {'size': 35}

# 그래프의 x축, y축으로 사용될 변수 설정
# sin, cos, tan 함수 사용
x = [i * 0.1 for i in range(-100, 101)]
y_sin = [math.sin(x[i]) for i in range(0, len(x))]
y_cos = [math.cos(x[i]) for i in range(0, len(x))]
y_tan = [math.tan(x[i]) for i in range(0, len(x))]

# 다중 그래프 작성, x축은 공유하되 y축은 공유하지 않음
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(20, 6), sharex=True, sharey=False)

# sin 그래프 작성
ax = axs[0]
ax.set_xlim(0, 10)  # x축의 범위 0 ~ 10
ax.set_ylim(-1.00, 1.00)  # y축의 범위 -1.00 ~ 1.00
ax.set_xticks(list(range(0, 11, 2)))  # x축의 눈금 0 ~ 10까지 2씩 나눠서 설정
ax.set_yticks([-1.00, -0.75, -0.50, -0.25, 0.00, 0.25, 0.50, 0.75, 1.00])  # y축의 눈금 리스트로 작성 (실수 단위)
ax.plot(x, y_sin, color='red', lw=5)  # 그래프 작성 / 색깔 : 빨간색, 굵기 : 5
ax.set_xlabel('x', fontdict=font)  # x축의 레이블 설정, 폰트는 기존의 작성한 font 딕셔너리 사용
ax.set_ylabel('sin(x)', fontdict=font)  # y축의 레이블 설정

# cos 그래프 작성
ax = axs[1]
ax.set_xlim(0, 10)  # x축의 범위 0 ~ 10
ax.set_ylim(-1.00, 1.00)  # y축의 범위 -1.00 ~ 1.00
ax.set_xticks(list(range(0, 11, 2)))  # x축의 눈금 0 ~ 10까지 2씩 나눠서 설정
ax.set_yticks([-1.00, -0.75, -0.50, -0.25, 0.00, 0.25, 0.50, 0.75, 1.00])  # y축의 눈금 리스트로 작성 (실수 단위)
ax.plot(x, y_cos, color='green', lw=5)  # 그래프 작성 / 색깔 : 초록색, 굵기 : 5
ax.set_xlabel('x', fontdict=font)  # x축의 레이블 설정, 폰트는 기존의 작성한 font 딕셔너리 사용
ax.set_ylabel('cos(x)', fontdict=font)  # y축의 레이블 설정

# tan 그래프 작성
ax = axs[2]
ax.set_xlim(0, 10)  # x축의 범위 0 ~ 10
ax.set_ylim(-600, 200)# y축의 범위 -600 ~ 200
ax.set_xticks(list(range(0, 11, 2)))  # x축의 눈금 0 ~ 10까지 2씩 나눠서 설정
ax.set_yticks(list(range(-600, 200, 200)))  # y축의 눈금 -600 ~ 200
ax.plot(x, y_tan, color='blue', lw=5)  # 그래프 작성 / 색깔 : 파란색, 굵기 : 5
ax.set_xlabel('x', fontdict=font)  # x축의 레이블 설정, 폰트는 기존의 작성한 font 딕셔너리 사용
ax.set_ylabel('tan(x)', fontdict=font)  # y축의 레이블 설정

# 3개의 그래프 사이의 간격 조정
plt.subplots_adjust(left=0.1, right=0.6, bottom=0.1, top=0.9, wspace=0.4, hspace=0.5)

#Q1 End
```

#### Q2.  Iris dataset은 붓꽃의 개체별 꽃받침, 꽃잎에 대한 길이, 너비 정보를 포함한다. 해당 데이터를 분석하고 꽃잎의 길이, 너비에 대해 히스토그램으로 시각화하시오.

##### 1. 꽃잎에 대한 길이, 너비를 hist(), subplots()으로 각각 시각화한다.
##### 2. 히스토그램을 나누는 구간 수는 적절히 선택한다. 
##### 3. 아래 예시와 같이 출력한다.
##### 예시:
![nn](fig_02.png)


```python
#Q2 Start
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from matplotlib import rc

# Iris 데이터셋 로드
iris = load_iris() 
data = iris.data # 컬럼별로 다음을 의미함. [sepal length, sepal width, petal length, petal width] 

# 다중 그래프 작성, x축, y축 공유하지 않음
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20, 8), sharex=False, sharey=False)
# 그래프에 사용될 y축 데이터
PetalLength = []
PetalWidth = []
# 폰트의 크기를 설정하기 위한 딕셔너리
font = {'size': 20}

# 반복문을 통해 Iris 데이터셋에서 데이터를 추출 후 리스트에 추가
for i in range(len(data)) :
    PetalLength.append(data[i][2])
    PetalWidth.append(data[i][3])
    
# PetalLength 그래프 작성
ax = axs[0]
ax.hist(PetalLength, bins=20, color = 'red', edgecolor='k')   #히스토리 20으로 설정, 그래프 색 : 빨간색, 그래프의 테두리 색 : 검은색
ax.set_title("Petal Length", fontdict=font)   # PetalLength 그래프의 제목 설정
ax.set_xlabel('Range(cm)', fontdict=font)     # x축의 레이블 설정, 폰트는 기존의 작성한 font 딕셔너리 사용 
ax.set_ylabel('Frequerency', fontdict=font)   # y축의 레이블 설정

# PetalWidth 그래프 작성
ax = axs[1]
ax.hist(PetalWidth, bins=20, color = 'blue', edgecolor='k')   #히스토리 20으로 설정, 그래프 색 : 파간색, 그래프의 테두리 색 : 검은색
ax.set_title("Petal Width", fontdict=font)   # PetalLength 그래프의 제목 설정
ax.set_xlabel('Range(cm)', fontdict=font)    # x축의 레이블 설정, 폰트는 기존의 작성한 font 딕셔너리 사용 
ax.set_ylabel('Frequerency', fontdict=font)  # y축의 레이블 설정

# 그래프의 크기 조정
plt.figure(figsize=(10, 6))  
plt.show()

#Q2 End
```

#### Q3. 한기대의 신입생 출신 지역 비율을 pie()를 통해 시각화하시오.
##### 1. 신입생의 수는 1,000명이며 랜덤으로 출신 지역을 배정한다.
##### 2. 아래 예시와 같이 출력한다. 
##### 예시:

![nn](fig_03.png)


```python
#Q3 Start
import matplotlib.pyplot as plt
import random

#지역명
regions = ['Seoul', 'Gyeonggi', 'Gangwon', 'Chungcheong', 'Gyeongsang', 'Jeolla', 'Jeju ']
#지역의 개수 만큼 리스트 선언
portion = [0] * len(regions)

# 총 인원
total_population = 1000
Total = 1000

# 반복문과 random를 이용하여 랜덤으로 출신 지역에 배치
for i in range(len(portion)):
    if i == len(portion) - 1:
        portion[i] = Total
    else:
        value = random.randint(1, Total)
        portion[i] = value
        Total -= value

# 그래프의 제목
plt.title('Hometown rate of Koreatech freshman')
# 파이 그래프 작성 / 음영 x, 그래프에서 빠져나오는 영역 x / 반시계 방향 / 소수점 첫재짜리까지 출력
plt.pie(portion, explode = (0, 0, 0, 0, 0, 0, 0), labels = regions, autopct='%1.1f%%', shadow=False, startangle=90)
plt.show()

#Q3 End
```

#### Q4. 주어진 데이터는 10월 넷째주 일주일 동안 측정된 평균 기온이다. 아래 요구에 맞게 시각화하시오. 

##### 1. 일주일 동안의 기온 그래프를 생성한다. Xlabel: 요일, Ylabel: 기온을 표시.
##### 2. 평균 기온을 계산하여 그래프에 표시한다. (점선으로 표기)
##### 3. 아래 예시와 같이 label을 지정한다.
##### 예시:

![nn](fig_04.png)


```python
#Q4 Start
import matplotlib.pyplot as plt
from matplotlib import rc

# 그래프의 x축, y축에 사용될 변수(리스트)
days = ['10.23', '10.24', '10.25', '10.26', '10.27', '10.28', '10.29']
temperatures = [15, 16, 13, 9, 12, 8, 7]
# 일주일 온도의 평균 값을 저장한 리스트
avg = [sum(temperatures) / len(temperatures) for _ in range(len(days))]

# 일주일 동안의 온도 그래프 출력 / 마크는 원 모양 / 범례 표시
plt.plot(days, temperatures, label = 'Temperature', marker = 'o')
# 일주일의 평균 온도 출력 / 그래프 색깔 : 빨간색 / 범례 표시 / 파선으로 출력
plt.plot(days, avg, color = 'red', label = 'Avg Temperature', linestyle='--')
plt.title('Temperature for a week')   # 그래프의 타이틀 설정
plt.xlabel('day')    # 그래프의 x축 제목 설정
plt.ylabel('Temperature (°C)')   # 그래프의 y축 제목 설정
plt.legend(loc='upper right')     # 범례 위치 조정
plt.show()

#Q4 End
```

#### Q5. Q4에서 작성한 그래프에 날짜 별 평균기온과의 차이를 bar()를 통해 시각화하시오.
##### 1. bar()는 0을 기준으로 형성되나 bottom 인자를 이용해 기준점을 설정할 수 있음.
##### 2. 아래 예시와 같이 출력한다.
##### 예시: 
![nn](fig_05.png)


```python
#Q5 Start
import matplotlib.pyplot as plt

# 그래프의 x축, y축에 사용될 변수(리스트)
days = ['10.23', '10.24', '10.25', '10.26', '10.27', '10.28', '10.29']
temperatures = [15, 16, 13, 9, 12, 8, 7]
# 일주일 온도의 평균 값을 저장한 리스트
avg = [sum(temperatures) / len(temperatures) for _ in range(len(days))]
# 각 일의 온도와 일주일 온도의 평균값을 저장한 리스트
gap = [temperatures[i] - avg[i] for i in range(len(days))]

# 그래프의 크기 조정
plt.figure(figsize=(20, 10))

# 일주일 동안의 온도 그래프 출력 / 마크는 원 모양 / 범례 표시
plt.plot(days, temperatures, label='Temperature', marker='o')
# 일주일의 평균 온도 출력 / 그래프 색깔 : 빨간색 / 범례 표시 / 파선으로 출력
plt.plot(days, avg, color='red', label='Avg Temperature', linestyle='--')
# 막대 그래프 출력 / 각 바 그래프의 시작지점을 avg 변수의 값으로 설정 / 범례 표시 / 투명도 설정
plt.bar(days, gap, bottom=avg[0], label='Temperature Difference', alpha=0.5)
plt.title('Temperature for a week')  # 그래프의 제목 설정
plt.xlabel('Day')    # 그래프의 x축 제목 설정
plt.ylabel('Temperature (°C)')    # 그래프의 y축 제목 설정
plt.legend(loc='upper right')      # 범례 위치 조정

plt.show()

#Q5 End
```

#### Q6. 가상의 세 기업의 월 매출 데이터를 랜덤하게 생성하고 이를 아래 요구에 맞게 시각화하시오. 

##### 1. 월 매출 데이터는 10000에서 30000 사이의 100단위 정수로 랜덤하게 생성한다. 

##### 2. 하나의 차트에 월 매출 데이터를 선 그래프로 표현한다. (마커, 선 스타일, 색상을 다르게 지정)

##### 3. 아래 예시와 같이 label을 지정하고 그리드를 표시한다.

##### 예시:

![nn](fig_06.png)

```python
#Q6 Start
import matplotlib.pyplot as plt
import random

# Data
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']

# 10000부터 30000까지 랜덤으로 정수를 발생시키고 이를 100단위 정수로 변환 후 리스트에 저장
sales_company_a = [random.randint(10000, 30000) // 100 * 100 for _ in range(len(months))]
sales_company_b = [random.randint(10000, 30000) // 100 * 100 for _ in range(len(months))]
sales_company_c = [random.randint(10000, 30000) // 100 * 100 for _ in range(len(months))]

# 각 그래프의 범례와 색깔, 마크 설정
plt.plot(months, sales_company_a, label='Company A', color='blue', marker='o')   # 그래프 색 : 파란색, 마크 원 모양
plt.plot(months, sales_company_b, label='Company B', color='green', marker='d', linestyle='--')  # 그래프 색 : 초록색, 마크 다이아몬드 모양, 파선으로 출력
plt.plot(months, sales_company_c, label='Company C', color='red', marker='s', linestyle='-.') # 그래프 색 : 빨간색, 마크 모양 사각형, 일점쇄선으로 출력

plt.xlabel('Months') # 그래프의 x축 제목 설정 
plt.ylabel('Sales ($)') # 그래프의 y축 제목 설정
plt.title('Monthly Sales Trends for Three Companies') # 그래프의 타이틀 설정
plt.grid(True)   # 그래프의 그리드 표시
plt.legend(loc='upper left')  # 범례 위치 조정

# 여백 조절하여 그래프를 가운데로 옮김
plt.figure(figsize=(10, 6)) 
plt.show()


#Q6 End
```
