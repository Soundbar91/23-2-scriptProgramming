#### Q1. 자동차를 표현하는 Car 클래스를 정의하고 자동차 객체를 생성하여 이용하는 프로그램을 작성하시오.
#### Car 객체를 생성하고 시동을 켜고 속도 증감을 수차례 수행 후 현재 속도와 연료량을 출력하여 테스트한다.

##### Car 클래스 정의 specification

- Car의 속성: 시동 여부, 속도, 연료량 
  - isOn, speed, fuel (모두 private 변수로 정의할 것. 속도, 연료량은 정수로 표현된다고 가정)

- Car의 동작:
  - 자동차 객체를 생성하는 생성자 (시동 여부, 속도, 연료량의 기본값은 False, 0, 100)
  - 각 변수에 대한 접근자 메소드
  - 시동을 켜는 start(self) 메소드
  - 시동을 끄는 stop(self) 메소드 – 시동이 켜져 있고, 속도가 0인 경우에만 시동을 끌 수 있음
  - 일정 단위만큼 속도를 증가하는 accel(self) 메소드 – 한번 호출 시마다 speed 10 증가, fuel 1 감소
  - 일정 단위만큼 속도를 감소하는 brake(self) 메소드 – 한번 호출 시마다 speed 20 감소
  
- 위에 정의되지 않은 부분에 대해서는 나름대로 논리적으로 결정하고 해당 사항을 주석으로 설명할 것

```python
#Q1 Start

# Car 클래스
class Car:

    # Car 클래스의 생성자
    # 디폴트 매개변수를 사용하여 기본값을 False, 0, 100으로 설정
    def __init__(self, isOn = False, speed = 0, fuel = 100) :
        # isOn, speed, fuel 모두 private 변수로 정의
        self.__isOn = isOn
        self.__speed = speed
        self.__fuel = fuel
    
    # Car 클래스의 멤버 변수에 대한 접근자 메소드
    # 반환 값으로 해당하는 값을 반환
    def getisOn(self) : return self.__isOn
    def getspeed(self) : return self.__speed
    def getfuel(self) : return self.__fuel

    # 시동을 켜는 메소드
    def start(self) : 
        # 연료가 없으면 시동을 킬 수 없으므로 메세지를 출력한다.
        if self.__fuel == 0 : print("연료가 없습니다.")
        # 시동이 켜져있는 상태이면 시동을 킬 수 없으므로 메세지를 출력한다.
        elif self.__isOn : print("시동이 켜져 있습니다.")
        # 시동을 킨다.
        else : self.__isOn = True
            

    # 시동을 끄는 메소드
    def stop(self) :
        # 시동이 켜져있으면 시동을 끈다.
        # 시동이 꺼져있는 상태이면 시동을 끌 수 없으므로 메세지를 출력한다.
        if self.__isOn : self.__isOn = False
        else : print("시동이 꺼져 있습니다.")

    # 속도를 증가하는 메소드
    def accel(self) :
        # 연료가 0이면 속도를 증가 시킬 수 없으므로 메세지를 출력한다.
        if (self.__fuel == 0) : print("연료가 없습니다.")
        # 시동이 꺼져 있으면 속도를 증가 시킬 수 없으므로 메세지를 출력한다.
        elif not self.__isOn : print("시동이 꺼져 있습니다.")
        # 연료가 충분하면 속도를 올리고 연료를 감소 시킨다.
        else :
            self.__speed += 10
            self.__fuel -= 1

    # 속도를 감소하는 메소드
    def brake(self) :
        # 속도를 감소 시켰을 때 속도가 음수가 되면 안되므로 메세지를 출력한다.
        if self.__speed - 20 < 0 : print("감속을 하면 속도가 0을 초과하게 됩니다.")
        # 시동이 꺼져 있으면 속도를 감소 시킬 수 없으므로 메세지를 출력한다.
        elif not self.__isOn : print("시동이 꺼져 있습니다.")
        # 속도를 감소 시킨다.
        else : self.__speed -= 20

# 시동이 켜져있고 속도가 10, 연료가 100인 Car 객체 변수를 선언
c = Car(True, 10, 100)
# 접근자를 통해 c의 상태를 출력
print("c의 시동 상태 :", c.getisOn())
print("c의 속도 :", c.getspeed())
print("c의 연로 :", c.getfuel())

# 반복문을 통해 속도를 가속
for i in range(10) :
    c.accel()
    print("가속 된 c의 속도 :", c.getspeed(), "\tc의 연료 :", c.getfuel())

print()

# 반복문을 통해 속도를 감속
for i in range(10) :
    c.brake()
    print("감속 된 c의 속도 :", c.getspeed())


#Q1 End
```

#### Q2. 엘리베이터를 표현하는 Elevator 클래스를 정의하고 엘리베이터 객체를 생성하여 이용하는 프로그램을 작성하시오.
#### Elevator 객체를 생성하고 엘리베이터 탑승, 하차 및 상하 이동을 수차례 수행한 후 현재 탑승 인원 및 층수를 출력하여 테스트한다.

##### Elevator 클래스 정의 specification

- Elevator의 속성: 현재 층, 탑승 인원
  - floor, people (모두 private 변수로 정의할 것)

- Elevator의 동작:
  - 엘리베이터 객체를 생성하는 생성자 (현재 층, 탑승 인원의 기본값은 1, 0)
  - 각 변수에 대한 접근자 메소드
  - 한 층 올라가는 up(self) 메소드 - 10층에서는 더 올라갈 수 없음
  - 한 층 내려가는 down(self) 메소드 – 1층에서는 내려갈 수 없음
  - 사람이 한 명 탑승하는 inPerson(self) 메소드 – 최대 인원 5명을 넘을 수 없음
  - 사람이 한 명 하차하는 outPerson(self) 메소드

- 위에 정의되지 않은 부분에 대해서는 나름대로 논리적으로 결정하고 해당 사항을 주석으로 설명할 것


```python
#Q2 Start

# Elevator 클래스
class Elevator:
    # Elevator 클래스의 생성자
    # 디폴트 매개변수를 이용하여 층과 탑승 인원의 기본 값을 1, 0으로 초기화
    def __init__ (self, floor = 1, people = 0) :
        # floor, people 멤버 변수를 private로 선언
        self.__floor = floor
        self.__people = people

    # private 멤버 변수에 대한 접근자 메소드
    # 반환 값으로 해당 변수의 값을 반환
    def getfloor(self) : return self.__floor
    def getpeople(self) : return self.__people

    # 한 층 올라가는 메소드
    def up(self) :
        # 현재 층수가 10층이면 메세지를 출력한다.
        if self.__floor == 10 : print("더 이상 올라갈 수 없습니다.")
        # 한 층 올라간다.
        else : self.__floor += 1

    # 한 층 내려가는 메소드
    def down(self) :
        # 현재 층수가 1층이면 메세지를 출력한다.
        if self.__floor == 1 : print("더 이상 내려갈 수 없습니다.")
        # 한 층 내려간다.
        else : self.__floor -= 1

    # 사람이 한 명 탑승하는 메소드
    def inPerson(self) :
        # 탑승 인원이 5명이면 메세지를 출력한다.
        if self.__people == 5 : print("더 이상 탑승할 수 없습니다.")
        # 사람이 한 명 탑승한다.
        else : self.__people += 1
    
    # 사람이 한 명 하차하는 메소드
    def outPerson(self) :
        # 탑승 인원이 0명이면 메세지를 출력한다.
        if self.__people == 0 : print("하차 할 인원이 없습니다.")
        # 사람이 한 명 하차한다.
        else : self.__people -= 1

# 1층에 0명이 탑승하고 있는 Elevator 객체 변수 El를 선언
El = Elevator(1, 0)
# 현재 탑승 인원과 현재 층을 접근자를 통해 출력
print("현재 탑승 한 인원 :", El.getpeople(), "\t현재 층 :", El.getfloor())

# 반복문을 통해 층 수를 올리고, 사람을 탑승 시킨다.
for i in range(5) :
    El.up()
    El.inPerson()
    print("현재 탑승 한 인원 :", El.getpeople(), "\t현재 층 :", El.getfloor())

print()

# 반복문을 통해 층 수를 내리고, 사람을 탑승 시킨다.
for i in range(2) :
    El.down()
    El.inPerson()
    print("현재 탑승 한 인원 :", El.getpeople(), "\t현재 층 :", El.getfloor())

print()

# 반복문을 통해 층 수를 내리고, 사람을 하차 시킨다.
for i in range(2) :
    El.down()
    El.outPerson()
    print("현재 탑승 한 인원 :", El.getpeople(), "\t현재 층 :", El.getfloor())

print()

# 층수를 내리고 사람을 하차 시킨다.
El.down()
for i in range(3) : El.outPerson()
print("현재 탑승 한 인원 :", El.getpeople(), "\t현재 층 :", El.getfloor())
El.outPerson()


#Q2 End
```

#### Q3. 제품을 표현하는 Product 클래스를 정의하고 객체를 생성하여 이용하는 프로그램을 작성하시오. 

#### Product 객체를 여러 개 생성하고 가격 업데이트, 재고량 증가, 제품 판매, 제품 재고 확인, 제품 정보 출력 등의 기능을 수행하여 테스트한다.

##### Product 클래스 정의 specification

- Product의 속성: 제품명, 제품 가격, 제품 재고량
  - name, price, quantity (모두 private 변수로 정의할 것)
  
- Product의 동작:
  - 제품 객체를 생성하는 생성자 (인스턴스 변수를 초기화 하기 위해 3개의 매개 변수를 정의)
  - 각 인스턴스 변수에 대한 접근자
  - 제품 가격을 업데이트 하는 setPrice(self, price) 메소드
  - 제품 재고량을 증가하는 increaseQuantity(self, amount) 메소드
  - 제품을 판매하고 재고량을 감소하는 sell(self, quantitySold) 메소드 - 남은 재고량보다 많이 판매할 수 없음
  - 제품 재고를 확인하여 일정 수준 미만으로 떨어지면 재고 보충 알림 메시지를 출력하는 checkStockLevel(self, threshold=10) 메소드
  - 제품에 할인을 적용하는 applyDiscount(self, discountRate) 메소드 - 할인율(DiscountRate)만큼 제품 가격을 감소
  - 제품 정보를 출력하는 getProductInfo(self) 메소드 - 제품명, 가격, 재고량 순서대로 읽기 쉬운 형태로 화면에 출력

- 위에 정의되지 않은 부분에 대해서는 나름대로 논리적으로 결정하고 해당 사항을 주석으로 설명할 것

```python
#Q3 Start

# Product 클래스
class Product:

    # Prduct 클래스의 생성자
    # 디폴트 매개변수를 이용하여 상품명, 가격, 재고량을 초기화
    def __init__ (self, name = "NULL", price = 0, quantity = 0) :
        # 멤버 변수 name, price, quantity를 private로 선언
        self.__name = name
        self.__price = price
        self.__quantity = quantity

    # 멤버 변수에 대한 접근자
    # 반환 값으로 해당하는 멤버 변수의 값을 반환
    def getname(self) : return self.__name
    def getprice(self) : return self.__price
    def getquantity(self) : return self.__quantity

    # 제품 가격을 업데이트 하는 메소드
    # 매개변수로 전달받은 값으로 가격을 최신화
    def setPrice(self, price) :
        self.__price = price

    # 제품 재고량을 증가하는 메소드
    # 매개변수로 전달받은 값만큼 재고량을 증가
    def increaseQuantity(self, amount) :
        self.__quantity += amount

    # 제품을 판매하는 메소드
    # 매개변수로 전달받은 값만큼 재고량을 감소
    def sell(self, quantitySold) : 
        # 판매할려는 양보다 재고량이 부족하면 메세지를 출력
        if self.__quantity < quantitySold : print("재고가 부족합니다.")
        # 재고량을 감소 시킨다.
        else : self.__quantity -= quantitySold

    # 재고량을 확인하는 메소드
    # 매개변수로 아무 값도 전달이 되지 않으면 기본 값 10을 기준으로 재고가 부족한지 확인
    def checkStockLevel(self, threshold = 10) :
        # 재고량이 일정 수준 미만인 경우 메세지를 출력
        if self.__quantity < threshold : print("재고를 보충해야 합니다.")
        else : pass

    # 제품의 가격을 할인하는 함수
    # 매개변수로 전달받은 할인율 만큼 가격을 감소
    def applyDiscount(self, discountRate) :
        # 할인을 했을 때 가격이 0원이 되면 메세지를 출력
        if (self.__price - (self.__price * discountRate) == 0) : print("제품의 가격이 0원이 됩니다.")
        # 할인율 만큼 가격을 감소시킨다.
        else : self.__price -= self.__price * discountRate

    # 제품의 정보를 출력하는 함수
    # 접근자를 통해 정보를 출력
    def getProductInfo(self) :
        print("제품명 :", self.getname(), "제품 가격 :", self.getprice(), "재품 재고량 :", self.getquantity())

# 상품명은 삼다수, 가격은 1000원, 재고량은 5개인 Product 객체 변수 Water를 선언
Water = Product("삼다수", 1000, 5)
Water.getProductInfo()
print()

# 상품을 판매
print("삼다수 1개 판매")
Water.sell(1)
Water.getProductInfo()

# 상품의 재고 파악
print()
print("삼다수 재고 파악")
Water.checkStockLevel()

# 상품의 재고 보충
print()
print("삼다수 재고  10개 보충")
Water.increaseQuantity(10)
Water.getProductInfo()

# 상품 가격 할인
print()
print("삼다수 50% 할인")
Water.applyDiscount(0.5)
Water.getProductInfo()

#Q3 End
```
